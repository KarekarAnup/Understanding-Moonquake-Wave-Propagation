4. **Frequency Analysis**: The Fourier Transform is applied to the seismic signal to convert it from the time domain to the frequency domain. This analysis reveals the amplitude of different frequency components, helping to identify the dominant frequencies that characterize the seismic event. The power spectrum is plotted to visualize these frequency components.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.fft import fft

# Generate synthetic seismic data
np.random.seed(0)
duration = 10  # Duration of seismic data in seconds
sampling_rate = 1000  # Sampling rate in Hz
time = np.linspace(0, duration, duration * sampling_rate, endpoint=False)
frequency = 5  # Frequency of seismic signal in Hz
amplitude = 1  # Amplitude of seismic signal
earthquake_signal = amplitude * np.sin(2 * np.pi * frequency * time) + 0.5 * np.random.randn(len(time))

# Plot the synthetic seismic data
plt.figure(figsize=(10, 5))
plt.plot(time, earthquake_signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Synthetic Seismic Data (Simulated Earthquake)')
plt.grid(True)
plt.show()

# Basic analysis
max_amplitude = np.max(earthquake_signal)
print("Maximum Amplitude:", max_amplitude)

# Find peaks in the seismic signal
peaks, _ = find_peaks(earthquake_signal, height=0.5)

# Plot the synthetic seismic data with detected peaks
plt.figure(figsize=(10, 5))
plt.plot(time, earthquake_signal, label='Seismic Data')
plt.plot(time[peaks], earthquake_signal[peaks], 'r.', markersize=10, label='Detected Peaks')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Synthetic Seismic Data with Detected Peaks')
plt.legend()
plt.grid(True)
plt.show()

# Compute power spectrum using Fourier transform
freq = np.fft.fftfreq(len(time), d=1/sampling_rate)
spectrum = np.abs(np.fft.fft(earthquake_signal))

# Plot power spectrum
plt.figure(figsize=(10, 5))
plt.plot(freq[:len(freq)//2], spectrum[:len(freq)//2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Power Spectrum of Synthetic Seismic Data')
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Generate synthetic moonquake seismic data
np.random.seed(0)
duration = 10  # Duration of seismic data in seconds
sampling_rate = 1000  # Sampling rate in Hz
time = np.linspace(0, duration, duration * sampling_rate, endpoint=False)
frequency = 0.5  # Frequency of moonquake signal in Hz
amplitude = 1  # Amplitude of moonquake signal
moonquake_signal = amplitude * np.sin(2 * np.pi * frequency * time) + 0.5 * np.random.randn(len(time))

# Compute power spectrum using Fourier transform
freq = fftfreq(len(time), d=1/sampling_rate)
spectrum = np.abs(fft(moonquake_signal))

# Plot power spectrum
plt.figure(figsize=(10, 5))
plt.plot(freq[:len(freq)//2], spectrum[:len(freq)//2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Power Spectrum of Synthetic Seismic Data for Moonquake')
plt.grid(True)
plt.show()

# Dominant frequency identification
dominant_frequency_index = np.argmax(spectrum[:len(freq)//2])
dominant_frequency = freq[dominant_frequency_index]
print("Dominant Frequency:", dominant_frequency, "Hz")

# Frequency band analysis
low_frequency_band = spectrum[(freq >= 0) & (freq < 0.25)]  # Example: Low-frequency band
mid_frequency_band = spectrum[(freq >= 0.25) & (freq < 0.75)]  # Example: Mid-frequency band
high_frequency_band = spectrum[(freq >= 0.75) & (freq < 1)]  # Example: High-frequency band
print("Low Frequency Band Energy:", np.sum(low_frequency_band))
print("Mid Frequency Band Energy:", np.sum(mid_frequency_band))
print("High Frequency Band Energy:", np.sum(high_frequency_band))
````
